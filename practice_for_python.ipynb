```python
# The task has been completed already


# # Assuming 'hotel' is your DataFrame and 'Total Revenue' is the name of the column

# Remove comma and convert to integer
hotel['Total Revenue'] = hotel['Total Revenue'].str.replace(
    ',', '').astype(int)
```

```python
hotel['Total Revenue'] = hotel['Total Revenue'].astype(int)
```

# **Sorting the dataset for Top 10 Countries in Revenue and Room Nights**


```python
top10_rev = hotel.sort_values('Total Revenue', ascending=False).head(10)
top10_rev.fillna(0)
top10_rev
```

```python
top10_room = hotel.sort_values('Total Room Nights', ascending=False).head(10)
top10_room
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(top10_rev['Country Name'], top10_rev['Total Revenue'])
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.title('Top 10 Countries by Total Revenue in million(Thb)')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(top10_room['Country Name'], top10_rev['Total Room Nights'])
plt.xlabel('Country')
plt.ylabel('Total Room Nights')
plt.title('Top 10 Countries by Total Room Nights')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with columns 'Country Name' and 'Total Revenue'

# Sort the DataFrame by revenue in descending order
df_sorted = hotel.sort_values('Total Revenue', ascending=False)

# Extract the data for plotting
countries = df_sorted['Country Name'].astype(str).tolist()
revenue = df_sorted['Total Revenue'].tolist()

# Plot the line graph
plt.figure(figsize=(100, 24))
plt.plot(countries, revenue, marker='o')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.title('Revenue by Country')
plt.xticks(rotation=90)
plt.show()
```
